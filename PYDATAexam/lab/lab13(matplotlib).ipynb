```python
from matplotlib import font_manager, rc
font_path = "data/THEdog.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
# 문제1
data = [[500, 450, 520, 610], 
        [690, 700, 820, 900], 
        [1100, 1030, 1200, 1380], 
        [1500, 1650, 1700, 1850], 
        [1990, 2020, 2300, 2420], 
        [1020, 1600, 2200, 2550]]
index = ['2015', '2016', '2017', '2018' , '2019' , '2020']
columns = ['1분기', '2분기', '3분기', '4분기']
df = pd.DataFrame(data, index, columns)
display(df)
```

```python
# 문제2
plt.figure(figsize=(10,6))
plt.plot(df.index, data) # 다른방법: plt.plot(df.index, df.values)
# plt.xticks(index, labels=['2015년', '2016년', '2017년', '2018년', '2019년', '2020년'])
plt.xticks(index, df.index.map(lambda x : x+'년')) #방법2: lambda 사용
plt.legend(columns, loc='upper left')

title_font = {
    'fontsize': 20,
    'fontweight': 'bold',
    'color': 'purple'
}
plt.title('시각화 과제1', fontdict=title_font )
plt.savefig('output/hw1.png')

plt.show()
```

```python
# 문제3
plt.figure(figsize=(10,6))
df_t = df.T
plt.plot(df_t.index, df_t.values)
plt.legend(df_t.columns, loc='upper left')

title_font = {
    'fontsize': 20,
    'fontweight': 'bold',
    'color': 'red'
}
plt.title('시각화 과제2', fontdict=title_font )
plt.savefig('output/hw2.png')

plt.show()
```

```python
# 문제4
plt.figure(figsize=(10,6))

# ratio = [29.7, 20.8, 22.8, 26.7]

labels = df.columns
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['red', 'green', 'darkred', 'purple']

plt.pie(df.sum(), labels=labels, autopct='%.1f%%', explode=explode, shadow=True, colors=colors)
plt.savefig('output/hw3.png')

plt.show()
```
