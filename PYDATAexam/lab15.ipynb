**1. 다음에 제시된 코드를 수행시켜서 누락데이터를 일부 갖게 되는 데이터 프레임을 생성한다.**

```python
import pandas as pd
import seaborn as sns
import numpy as np
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']]
import random

random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
df1 = iris_x.head(10)
df = iris_x
```

```python
# (1)
df.dropna(how='any').head(3)
```

```python
# (2)
df.dropna(thresh=2).head(10)
```

```python
# (3)
df.dropna(subset=['sepal_length', 'sepal_width']).head(10)
```

```python
# (4) NaN -> 0으로 채우기
df.fillna(0, inplace=True)
df.head(10)
```

```python
# (5) NaN -> 평균으로 채우기
mean_sepal_length = round(df['sepal_length'].mean(axis=0),1)
mean_sepal_width = round(df['sepal_width'].mean(axis=0),1)
mean_petal_length = round(df['petal_length'].mean(axis=0),1)
mean_petal_width = round(df['petal_width'].mean(axis=0),1)
df['sepal_length'].fillna(mean_sepal_length)
df['sepal_width'].fillna(mean_sepal_width)
df['petal_length'].fillna(mean_petal_length)
df['petal_width'].fillna(mean_petal_width)
df.head(10)
```

```python
# (6) NaN -> 10으로 채우기
df.fillna(10, inplace=True)
df.head(10)
```

```python
# (7) NaN -> 근처값으로 채우기
df['sepal_length'].fillna(method='ffill', inplace=True)
df['sepal_width'].fillna(method='ffill', inplace=True)
df['petal_length'].fillna(method='ffill', inplace=True)
df['petal_width'].fillna(method='ffill', inplace=True)
df.head(10)
```

**3. product_click.log 파일을 읽어서 판다스의 DataFrame 객체로 생성하고 다음에 제시된 결과화면을 보면서 문제들을 해결해 본다.**

```python
# (1)
df = pd.read_csv('data/product_click.log', header=None, sep=' ')
print(df.head())
df.info()
```

```python
# (2)
df.columns = ['oldclicktime','clickped']
print(df.head())
```

```python
# (3)
df['oldclicktime'] = df['oldclicktime'].astype('str')
df.info()
```

```python
# (4)
# df['newclicktime'] = pd.to_datetime(df['oldclicktime'], format="%Y%m%d%H%M")
# df['newclicktime']
df['newclicktime'] = pd.to_datetime(df['oldclicktime'])
df['newclicktime']

df.info()
print(df.head())
```

```python
# (5)
df['year'] = df['newclicktime'].dt.year
df['month'] = df['newclicktime'].dt.month
df['day'] = df['newclicktime'].dt.day
df['hour'] = df['newclicktime'].dt.hour
df['minute'] = df['newclicktime'].dt.minute
df.info()
display(df.head())
```
