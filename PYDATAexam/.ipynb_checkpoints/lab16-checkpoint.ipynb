# 문제1

```python
import pandas as pd
df = pd.read_csv('data/cctv_seoul.csv')
df.head()
```

```python
def divide_two_obj(a, b):
    return a / b

df['CCTV비율'] = df.apply(lambda x: divide_two_obj(x['CCTV수'], x['인구수']), axis=1) * 100
df['노인비율'] = df.apply(lambda x: divide_two_obj(x['고령자'], x['인구수']), axis=1)
df.head()
```

# 문제2

```python
import pandas as pd
df2 = pd.read_csv('data/product_click.log', header=None, sep=' ')
df2.info()
df2.columns = ['시간', '상품 ID']
df2.head()
```

```python
df2['상품 ID'].value_counts()
```

```python
df2['시간'] = df2['시간'].astype('str')
df2['시간_new'] = pd.to_datetime(df2['시간'])
df2
```

```python
df2['시간'] = df2['시간_new'].dt.hour
df2['요일'] = df2['시간_new'].dt.dayofweek
df2
```

```python
# (2)
pv = pd.pivot_table(df2,
                     index=['시간'],
                     values='상품 ID',
                     aggfunc='count')
pv
```

```python
# (3)
pv = pd.pivot_table(df2,
                     index=['요일'],
                     values='상품 ID',
                     aggfunc='count')
pv.index
```

# 문제3

```python
df3 = pd.read_csv('data/emp.csv')
```

```python
# (1)
pdf3 = pd.pivot_table(df3,
                     index='deptno',
                     values='sal',
                     aggfunc='sum')
pdf3                    
```

```python
# (2)
pdf3 = pd.pivot_table(df3,
                     index='job',
                     values='sal',
                     aggfunc='sum')
pdf3                    
```

```python
# (3) - 가장 오래된 직원의 입사날짜
pdf3 = pd.pivot_table(df3,
                     index=['deptno', 'job'],
                     values=['sal', 'hiredate'],
                     aggfunc=['max'])
pdf3                    
```

```python
# (4)
pdf3 = pd.pivot_table(df3,
                     index=['job', 'deptno'],
                     values='sal',
                     aggfunc='max')
pdf3                    
```

```python
# (5)
pv1 = pd.pivot_table(df3,
                     index='deptno',
                     columns='job',
                     values='sal',
                     aggfunc='sum')
pv1
```

```python
# (6)
pv1.fillna(0)
```

```python
# (7)
import seaborn as sns
sns.heatmap(pv1, annot=True, cmap="PuRd", fmt='.1f')
```

```python
# (8)
df3['hiredatenew'] = pd.to_datetime(df3['hiredate'])
df3['year'] = df3['hiredatenew'].dt.year
df3['month'] = df3['hiredatenew'].dt.month
df3.info()
df3.head(5)
```

```python
# (9)
pv2 = pd.pivot_table(df3,
                     index=['year','month'],
                     columns='deptno',
                     values='sal',
                     aggfunc='max')
pv2 = pv2.fillna(0)
pv2
```

```python
# (10) - 1) 1981년 12월 데이터들을 추출한다.
print(pv2.xs((1981, 12)))
```

```python
# (10) - 2) 1982년 데이터들을 추출한다.
pv2.xs(1982)
```

```python
# (10) - 3) 매년 12월 데이터들을 추출한다.
pv2.xs(12, level='month')
```

```python
# (11)
sns.heatmap(pv2, cmap="PuRd", fmt='.1f')
```
